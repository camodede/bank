```python
documents = [
    {"type": "passport", "number": "2207 876234", "name": "Василий Гупкин"},
    {"type": "invoice", "number": "11-2", "name": "Геннадий Покемонов"},
    {"type": "insurance", "number": "10006", "name": "Аристарх Павлов"}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

#задача 1.1

def p_document(number):
    for doc_number in documents:
        if doc_number["number"] == number:
            print(' Владелец документа: ', doc_number["name"])
            break
    else:
        print(' Документ не найден в базе')
        
        
# Задача 1.2

def s_document(number):
    for s_directories in directories.items():
        for doc_number in s_directories[1]:
            if doc_number == number:
                print(' Документ хранится на полке: ', s_directories[0])
            break             
    else:
        print(' Документ не найден в базе')
        
# Задача 1.3

def all_doc():
    for do in documents:
        print(f"№:{do['number']},тип: {do['type']}, владелец: {do['name']},Полка хранения,{s_document(do['name'])}")
    
   
    
# Задача 1.4

def as_document(new_doc):
    if  int(new_doc) == 1 or int(new_doc) == 2 or int(new_doc) == 3:
        print('Такое уже есть')
    else:
        new_as = {new_doc:[]}
        directories.update(new_as)
        print(f'Полка добавлена. Текущий перечень полок:{",".join(list(directories.keys()))}')
        
# Задача 1.5        

def ds_shelf(shelf):
    if len(directories[shelf]) == 0:
        del(directories[shelf])
        print(f'Полка удалена. Текущий перечень полок:{",".join(list(directories.keys()))}')
    if len(directories[shelf]) > 0:
        print(f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:{",".join(list(directories.keys()))}')
        return


def p():
    while True:
        command = input('Введите команду: ')
        if command == 'p':
            p_document(input('Введите номер документа:'))
            
        elif command == 's':
            s_document(input('Введите номер документа:'))
        elif command == 'l':
            all_doc()
        elif command == 'as':
            as_document(input('Введите номер полки:'))
        elif command == 'ds':
            ds_shelf(input('Введите номер полки:'))
            
        
        else:
            print('Вы ввели команду не корректно, повторите ввод.')
            

p()





        
```
